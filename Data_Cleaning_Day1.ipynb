```python
# Standardize 'gender' column
df['Gender'] = df['Gender'].str.strip().str.lower()


```

```python
df.head()
```

```python
# Standardize 'country' column
df['Country'] = df['Country'].str.strip().str.title()
```

```python
df.head()
```

```python
# Rename columns: lowercase and replace spaces with underscores
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
df.head()
```

```python
# Convert 'age' to integer (int8 for memory efficiency)
df['age'] = df['age'].astype('int8')

# Convert 'purchase_amount' to float (if not already)
df['purchase_amount'] = df['purchase_amount'].astype('float64')

# Convert 'purchase_date' to datetime
df['purchase_date'] = pd.to_datetime(df['purchase_date'], errors='coerce')
```
